```python
import pandas as pd
import pydotplus 
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
data=pd.DataFrame({'P_Movies':[17,64,18,20,38,49,55,25,29,31,33],'Gender':[1,0,1,0,1,0,0,1,1,0,1]})
data=data.sort_values('P_Movies')
```

```python
def tree_graph_to_png(tree, feature_names, png_file_to_save):
 tree_str = export_graphviz(tree, feature_names=feature_names,filled=True, out_file=None)
 graph = pydotplus.graph_from_dot_data(tree_str)
 graph.write_png(png_file_to_save)
```

```python
dt = DecisionTreeClassifier(criterion = 'entropy')
```

```python
#X is the features in this dataset
X = data['P_Movies'].values.reshape(-1, 1)
#Y is the vector with our Target Variables
Y = data['Gender'].values
```

```python
dt.fit(X, Y)
tree_graph_to_png(dt, feature_names=['P_Movies'],
png_file_to_save='dt.png')
d = np.array([7, 15, 43, 45])
d=d.reshape(-1, 1)
dt.predict(d)
```
